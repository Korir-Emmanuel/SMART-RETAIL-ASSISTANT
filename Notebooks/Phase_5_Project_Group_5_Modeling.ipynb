**Function to handle text processing**

```python
def text_pattern_check(process, pattern):
    def extract_pattern(text):
        if isinstance(text, str):
            return re.findall(pattern, text)
        return []

    # apply pattern extraction
    links_list = df['Item Description Clean'].apply(extract_pattern)

    # flatten the list of lists
    all_links = [link for sublist in links_list for link in sublist]

    # check if anything was found
    if all_links:
        print(f"{process} found in the dataset: {len(all_links)} matches")
        print(all_links[:5])  # optional: show a few samples
    else:
        print(f"No {process} found.")
```

- Replaced abreviation with its full form so as they won't be removed as stopwords

**Removing pattern function**

```python
def remove_pattern(pattern):
    df["Item Description Clean"] = df["Item Description Clean"].apply(lambda x: re.sub(pattern, '', x))
```

**Replacing abbreviation with its full form**

```python
# loading abbreviation dictionary from abbr.txt
with open('abbr.txt', 'r') as file:
    abbr_dict = json.load(file)
```

```python
def check_abbr():
    # Function to check if a text contains any abbreviation
    def contains_abbr(text):
        return any(word in abbr_dict for word in text.split())

    # Filter texts that contain abbreviations
    abbr_text = df[df["Item Description Clean"].apply(contains_abbr)]

    # Check if any were found and print message + examples
    if not abbr_text.empty:
        print("Abbreviations found in some texts. Here are a few examples:")
        print(abbr_text[["Item Description Clean"]].head(5))
    else:
        print("No abbreviations found in any texts.")
check_abbr()
```

```python
# expand abbreviations in a tweet
def expand_abbr_in_tweet(text):
    return " ".join([abbr_dict.get(word, word) for word in text.split()])

# apply the function to each tweet in the DataFrame
df["Item Description Clean"] = df["Item Description Clean"].apply(expand_abbr_in_tweet)
```

```python
# getting stopwords
stopwords_list = stopwords.words('english')

#removing punctuation from stopwords since punctuation have been removed
stopwords_list= [word.translate(str.maketrans('', '', string.punctuation)) for word in stopwords_list]
```

**SPEND FORECASTING WITH TIME SERIES.**

```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset again specifically for time series analysis
df_ts = pd.read_excel("Working Data.xlsx")

# Dropping columns with more than 90% missing values
df_ts = df_ts.loc[:, df_ts.isnull().mean() < 0.9]

# Dropping the 'Supplier Site' column
df_ts = df_ts.drop(columns=['Supplier Site'])

# Dropping rows with missing essential information for time series
df_ts = df_ts.dropna(subset=['UOM', 'Unit Price', 'Distribution Ordered Quantity'])

# Filling missing values in 'Department' and 'Division'
df_ts['Department'] = df_ts['Department'].fillna('Unknown')
df_ts['Division'] = df_ts['Division'].fillna('Unknown')

# 'Month' column contains date information
df_ts['Month'] = pd.to_datetime(df_ts['Month'], errors='coerce')

# Dropping rows where date conversion failed
df_ts = df_ts.dropna(subset=['Month'])

# time series analysis using df_ts
df_ts['ds'] = df_ts['Month']
monthly_spend = df_ts.groupby('ds')['PO Amount'].sum().reset_index()
monthly_spend.columns = ['ds', 'y']


model = Prophet(yearly_seasonality=True)
model.fit(monthly_spend)

# future- dataframe for forecasting
future = model.make_future_dataframe(periods=12, freq='M') # Forecast for the next 12 months

# Predict
forecast = model.predict(future)

# vizz.
model.plot(forecast)
plt.title('Monthly Spend Forecast with Yearly Seasonality')
plt.xlabel('Date')
plt.ylabel('Spend')
plt.show()

# Plot components to see seasonality
model.plot_components(forecast)
plt.show()
import warnings
warnings.filterwarnings("ignore")
```

**ANOMALY DETECTION USING ISOLATION FOREST.**

```python
from sklearn.ensemble import IsolationForest
import numpy as np
import pandas as pd # Import pandas if not already imported in this cell

# Reloading the dataset for anomaly detection
df_anomaly = pd.read_excel("Working Data.xlsx")

# Application of initial cleaning steps
df_anomaly = df_anomaly.loc[:, df_anomaly.isnull().mean() < 0.9]
df_anomaly = df_anomaly.drop(columns=['Supplier Site'])
df_anomaly = df_anomaly.dropna(subset=['UOM', 'Unit Price', 'Distribution Ordered Quantity'])
df_anomaly['Department'] = df_anomaly['Department'].fillna('Unknown')
df_anomaly['Division'] = df_anomaly['Division'].fillna('Unknown')

# Identifying numerical columns for anomaly detection after cleaning
numerical_cols_anomaly = ['PO Amount', 'Quantity', 'Unit Price']


if all(col in df_anomaly.columns for col in numerical_cols_anomaly):

    for col in numerical_cols_anomaly:
         df_anomaly[col] = np.log1p(df_anomaly[col])


    print("Log transformation applied successfully to the numerical columns in df_anomaly.")
else:
    print(f"Error: Some of the required numerical columns {numerical_cols_anomaly} are missing in df_anomaly.")


X_anomaly = df_anomaly[numerical_cols_anomaly]
model_iso = IsolationForest(contamination='auto', random_state=42)
df_anomaly['anomaly_score'] = model_iso.fit_predict(X_anomaly)
```

```python
# Calculate the anomaly scores
df_anomaly['raw_anomaly_score'] = model_iso.decision_function(X_anomaly)

# Visualize the distribution of scores
plt.figure(figsize=(8, 6))
sns.histplot(df_anomaly, x='raw_anomaly_score', hue='anomaly_score', kde=True)
plt.title('Distribution of Anomaly Scores')
plt.xlabel('Raw Anomaly Score')
plt.ylabel('Frequency')
plt.show()

# Examine the scores for identified anomalies
print("Raw anomaly scores for detected anomalies (-1):")
print(df_anomaly[df_anomaly['anomaly_score'] == -1]['raw_anomaly_score'].head())

# Examine the scores for identified inliers
print("\nRaw anomaly scores for detected inliers (1):")
print(df_anomaly[df_anomaly['anomaly_score'] == 1]['raw_anomaly_score'].head())
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
df_cluster = pd.read_excel("Working Data.xlsx")
df_cluster = df_cluster.loc[:, df_cluster.isnull().mean() < 0.9]
df_cluster = df_cluster.drop(columns=['Supplier Site'])
df_cluster = df_cluster.dropna(subset=['UOM', 'Unit Price', 'Distribution Ordered Quantity'])
df_cluster['Department'] = df_cluster['Department'].fillna('Unknown')
df_cluster['Division'] = df_cluster['Division'].fillna('Unknown')
features = df_cluster[['PO Amount', 'Quantity', 'Unit Price']].fillna(0)
scaled = StandardScaler().fit_transform(features)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_cluster['Cluster'] = kmeans.fit_predict(scaled)
print("Representative POs per cluster (mean values):")
print(df_cluster.groupby('Cluster')[['PO Amount', 'Quantity', 'Unit Price']].mean())
```
